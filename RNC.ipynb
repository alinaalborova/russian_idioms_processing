```python
morph = pymorphy2.MorphAnalyzer()
vncs_inflected = []
vncs_initial_forms = []
for vnc in vncs_list:
  vnc_list = vnc.split()
  verb = vnc_list[0]
  word = morph.parse(verb)[0]
  forms = word.lexeme[:13]
  if len(forms) < 13:
    print(len(forms), verb)
  forms_clean = [] 
  new_texts = []
  for form in forms:
    new_text = ''
    forms_clean.append(form[0])
    new_text += form[0] + ' ' + ' '.join(vnc_list[1:])
    new_texts.append(new_text)
    vncs_inflected.append(new_text)
    vncs_initial_forms.append(vnc)
print(vncs_inflected, len(vncs_inflected))
print(vncs_initial_forms, len(vncs_initial_forms))
```

## All Together

```python
results = rus_corp.search('бросали тень', n_results=10)
print('Res;', results)
for result in results:
    for i, target in enumerate(result):
        print(i+1, target.text)
        N = i + 1
    print(N)
results
```

```python
contexts = []
num_of_contexts = []
#N_for_one = []
for idiom in tqdm(vncs_inflected): 
  results = rus_corp.search(idiom, n_results=10)
  if results:
    for result in results:
      for i, target in enumerate(result):
        contexts.append(target.text)
        # num of contexts for a particular form
        N = i + 1
      num_of_contexts.append(N)
  else:
    N = 0
    num_of_contexts.append(N)
```

```python
all_vncs_by_N = []
for index, idiom in enumerate(vncs_inflected):
  for i in range(num_of_contexts[index]):
    all_vncs_by_N.append(idiom)
print(all_vncs_by_N[:20])
len(all_vncs_by_N)
```

```python
all_vncs_by_N_norm = []
for index, idiom in enumerate(vncs_inflected):
  for i in range(num_of_contexts[index]):
    all_vncs_by_N_norm.append(vncs_initial_forms[index])
print(all_vncs_by_N_norm[:20])
len(all_vncs_by_N_norm)
```

```python
vnc_contexts = pd.DataFrame(columns=['Idiom Normal', 'Idiom Inflected', 'Example'])
vnc_contexts['Idiom Normal'] = all_vncs_by_N_norm
vnc_contexts['Idiom Inflected'] = all_vncs_by_N
vnc_contexts['Example'] = contexts
vnc_contexts
```

```python
vnc_contexts.to_csv('vnc_contexts_inflected.csv', encoding='utf-8-sig')
```
